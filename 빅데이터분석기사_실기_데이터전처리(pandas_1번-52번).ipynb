```python
# [44]
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv'
DF = pd.read_csv(DataUrl)

DF.head(3)
```

```python
# [45-1]데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라 ★★★
DF.groupby(['host_name'],as_index=False).size().sort_values(by='host_name', ascending=True).head(5)

# [45-2] 데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라 ★★★
DF.host_name.value_counts().sort_index()
# 이렇게 해도 되구나...! 그렇긴 하네
```

```python
# [46] 데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라.
#      빈도수 컬럼은 counts로 명명하라 ★★★ 
DF.groupby(['host_name']).size().to_frame().rename(columns={0:'counts'}).sort_values(by='counts', ascending=False).head(5)
# to_frame() 처음들어봣네...!
```

```python
# [47]
DF.groupby(['neighbourhood_group','neighbourhood'], as_index=False).size().head(3)
```

```python
# [48] neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값들을 출력하라
DF.groupby(['neighbourhood_group','neighbourhood'], as_index=False).size().sort_values(by=['neighbourhood_group','size'], ascending=[True,False]).drop_duplicates('neighbourhood_group', keep='first')

# [48] 
# 정답 neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값들을 출력하라
Ans= DF.groupby(['neighbourhood_group','neighbourhood'], as_index=False).size()\
                  .groupby(['neighbourhood_group'], as_index=False).max()
```

```python
# [49] neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하여라 ★★★
DF.groupby('neighbourhood_group')['price'].agg(['mean', 'var', 'max', 'min'])

# [49]
# 정답 neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하여라 ★★★
Ans = DF[['neighbourhood_group','price']].groupby('neighbourhood_group').agg(['mean','var','max','min'])

# 출력되는거 조금 다르기 하네
```
